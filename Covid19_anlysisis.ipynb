<h2>Overall Impact Across Countries</h2>

```python
top_deaths = df.sort_values('TotalDeaths', ascending=False).head(10)
top_deaths.plot(kind='bar', x='Country/Region', y='TotalDeaths', title='Top 10 Countries by COVID-19 Deaths', color='red')
```

```python
top_cases = df.sort_values('TotalCases', ascending=False).head(10)
top_cases.plot(kind='bar', x='Country/Region', y='TotalCases', title='Top 10 Countries by COVID-19 Cases', color='orange')
```

```python
top_cases = df.sort_values('TotalRecovered', ascending=False).head(10)
top_cases.plot(kind='bar', x='Country/Region', y='TotalRecovered', title='Top 10 Countries Recovered by COVID-19 Cases', color='green')
```

```python
df['DeathRate'] = (df['TotalDeaths'] / df['TotalCases']) * 100
df.sort_values('DeathRate', ascending=False).head(10).plot(kind='bar', x='Country/Region', y='DeathRate', title='Top 10 Countries by COVID-19 Death Rate')
```

<h2>Impact Across Continents</h2>

```python
df.groupby('Continent')['TotalDeaths'].sum().plot(kind='pie', autopct='%1.1f%%', title='COVID-19 Deaths by Continent')
```

```python
df.groupby('Continent')['TotalCases'].sum().plot(kind='pie', autopct='%1.1f%%', title='COVID-19 Cases by Continent')
```

```python
df.groupby('Continent')['TotalRecovered'].sum().plot(kind='pie', autopct='%1.1f%%', title='COVID-19 Recoveredby Continent')
```

```python
df['ActiveCases'] = df['TotalCases'] - (df['TotalDeaths'] + df['TotalRecovered'])
global_summary = df[['TotalRecovered', 'TotalDeaths', 'ActiveCases']].sum()
global_summary.plot(kind='pie', autopct='%1.1f%%', title='Global Case Outcomes')
```
